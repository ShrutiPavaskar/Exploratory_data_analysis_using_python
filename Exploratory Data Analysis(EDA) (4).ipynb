#                         Exploratory Data Analysis Using Python


OBJECTIVE    

    
  This project aims to analyze a dataset containing information about freelancing jobs and provide valuable
insights into the freelancing market. The dataset includes details such as job titles, categories, experience levels,
budgets, client information, and posting dates. By conducting exploratory data analysis, this project uncovers patterns
and trends within the dataset, helping stakeholders make informed decisions and better understand the freelancing 
landscape.


INTRODUCTION

   The rise of freelancing has transformed the way people work, offering flexible opportunities and a global talent pool. 
However, navigating the freelancing market can be challenging due to the vast amount of data available. This project focuses 
on analyzing a comprehensive dataset of freelancing jobs to gain insights into various aspects of the market.

Feature Description


The dataset for this project consists of the following features:
1. Title: The title of the job posting.
2. Category Name: The category to which the job belongs.
3. Experience: The required level of experience for the job.
4. Sub Category Name: The sub-category to which the job belongs.
5. Currency: The currency in which the budget is specified.
6. Budget: The budget allocated for the job.
7. Location: The location where the job is based.
8. Freelancer Preferred From: The preferred location of the freelancer.
9. Type: The type of job (e.g., full-time, part-time, hourly).
10. Date Posted: The date when the job was posted.
11. Description: A description of the job.
12. Duration: The expected duration of the job (if specified).
13. Client Registration Date: The registration date of the client.
14. Client City: The city of the client.
15. Client Country: The country of the client.
16. Client Currency: The currency preferred by the client.
17. Client Job Title: The job title of the client (if specified).


Problem Statement



The project aims to address the following questions:

1) What are the most common job categories in the freelancing market?
2) How do experience levels vary across different locations?
3) Which categories have the highest and lowest budgets?
4) Which type of job most preferable?
5) What is the distribution of freelancing jobs across different locations?
6) Which countries have the highest number of job postings?
7) What is the budget distribution based on experience level?
8) Which Sub categories have the highest and lowest budgets?

```python
df["Experience"].head()
```

```python
#remove unwanted symbols($,$$,$$$) from experience

df['Experience'] = df['Experience'].replace({'Entry ($)':'Entry', 'Intermediate ($$)': 'Intermediate','Expert ($$$)': 'Expert'})   # repalce value
df["Experience"].head()
```

```python
#conevert all Budget Value in one currency i.e "USD"

def Budget_USD(row):            
    if row['Currency'] == 'EUR':
        return row['Budget'] * 1.0956
    elif row['Currency'] == 'GBP':
        return row['Budget'] * 1.28
    else:
        return row['Budget']
```

```python
df['Budget'] = df.apply(Budget_USD, axis=1)
```

```python
df['Budget'].head()
```

```python
df['Currency'] = df['Currency'].replace({'EUR':'USD', 'GBP': 'USD',' USD':'USD'})
df['Currency'].nunique()
```

```python
df.head()
```

```python
df.describe() #statistical information of budget column
```

# Feature Extraction

```python
df[['Posted_day', 'Posted_time']] = df['Date Posted'].str.split(' ', 1, expand=True)
df[['Posted_day', 'Posted_time']]
```

```python
df=df.drop(columns=['Date Posted',"Posted_time"])
```

```python
df[['Posted_year','Posted_Month','Posted_Date']]=df["Posted_day"].str.split('-', 3, expand=True)
```

```python
df[['Posted_year','Posted_Month','Posted_Date']]=df[['Posted_year','Posted_Month','Posted_Date']].astype('int32')
```

```python
df=df.drop(columns=["Posted_Date"])
```

```python
df[['client_year','client_Month','client_Date']]=df['Client Registration Date'].str.split('-', 3, expand=True)
```

```python
df[['client_year','client_Month','client_Date']]=df[['client_year','client_Month','client_Date']].astype('int32')
```

```python
df=df.drop(columns=['Client Registration Date','client_Date'])
```

```python
df.head()
```

# Analysis Of Data

```python
category_count=df["Category Name"].value_counts()
category_count
```

```python
ex=[0.2,0,0,0,0,0,0,0,0]
plt.pie(category_count,explode=ex,labels=category_count.index,autopct="%0.3f%%",shadow=True)
plt.title("Category Distribution")
plt.show()
```

From above graph we understand Design has highest and Music and Audio has lowest distribution

```python
Experience_count=df["Experience"].value_counts()

ex=[0,0,0]
plt.pie(Experience_count,explode=ex,labels=Experience_count.index,autopct="%0.2f%%",shadow=True)
plt.title("Experience Distribution")
plt.show()
```

From above graph we see that Expert has highest disribution followed by Entry

```python
Category_Name_group=df.groupby("Category Name")["Budget"].sum()
sns.barplot(x=Category_Name_group,y=Category_Name_group.index,palette="rainbow")
```

```python
lowest = Category_Name_group.head(5).sort_values(ascending=False)
sns.barplot(x=lowest,y=lowest.index,palette="rainbow")
```

From above graph we see that Design, Business, Digital Marketing ,Marketing,Branding & Sales,Music need lowest Budget.

```python
Highest=Category_Name_group.tail(5).sort_values(ascending=False)
sns.barplot(x=Highest,y=Highest.index,palette="rainbow")
```

From above graph we see that Technology & Programming,Writing & Translation,Video,photo & Image,Social Media,Music & Audio need Highest budget.

```python
location_counts=df["Location"].value_counts()
plt.pie(location_counts,labels=location_counts.index,autopct='%0.2f%%',shadow=True)
plt.title("Location_Wise_Dist")
plt.show()
```

From above graph we see that laocation wise distribution remote location has highest demand.

```python
top=df["Client Country"].value_counts().sort_values(ascending=False)
top_new=top.head(5)
sns.barplot(x=top_new.values,y=top_new.index)
plt.title("top 5 Country_higher_job")
plt.xlabel("Country")
plt.ylabel("No of jobs")
```

From above graph we see that there is high jobs in United Kingdom Followed by United States, India, Spain, canada.

```python
Exp_group=df.groupby("Experience")["Budget"].sum()
ex=(0,0,0)
plt.pie(Exp_group,explode=ex,labels=Exp_group.index,autopct="%0.3f%%")
```
